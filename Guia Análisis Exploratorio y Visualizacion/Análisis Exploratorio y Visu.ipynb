```python
import numpy as np
import pandas as pd

ufosDf = pd.read_csv('Data/scrubbed.csv')
```

Luego de analizar los datos publicados en: https://www.kaggle.com/NUFORC/ufo-sightings Indicar cuál de las siguientes afirmaciones es correcta:	 

a) 	Gran Bretaña tiene mas reportes de avistajes que Canadá

b) 	Las Vegas, NV tiene un registro que aparece en el Top10 de avistajes con mayor duración en segundos para el mes de Marzo 2004.

c) 	La forma mas común en los ufos registrados es circular

d) 	El avistaje mas corto ocurrido en Marzo del 2007 ocurrió en la ciudad de Denver

```python
# a) Gran Bretaña tiene mas reportes de avistajes que Canadá.
ufosCountries = ufosDf[['country']]
ufosCountries.dropna(subset = ['country'], inplace = True)
ufosCountries['count'] = 1
ufosCountries = ufosCountries.groupby(['country'], as_index = False).count()
ufosCountries = ufosCountries.sort_values(by = 'count', ascending = False)
ufosCountries

# RTA: Falso - Gran Bretaña tiene 2050 avistajes y Canada 3266.
```

```python
# b) Las Vegas, NV tiene un registro que aparece en el Top10 de avistajes con mayor duración en segundos para el 
#    mes de Marzo 2004.
ufosLasVegas = ufosDf[['datetime', 'city', 'state', 'duration (seconds)']]

# Formateo la fecha y tomo el mes y el anio.
ufosLasVegas['datetime'] = pd.to_datetime(ufosLasVegas['datetime'], format='%m/%d/%Y %H:%M', errors='coerce')
ufosLasVegas.dropna(subset = ['datetime'], inplace = True)
ufosLasVegas['anio'] = ufosLasVegas['datetime'].apply(lambda x: x.strftime('%Y'))
ufosLasVegas['mes'] = ufosLasVegas['datetime'].apply(lambda x: x.strftime('%m'))

# Me quedo solo con los avistajes de marzo 2014.
ufosLasVegas = ufosLasVegas.loc[ufosLasVegas['anio'] == '2004', :]
ufosLasVegas = ufosLasVegas.loc[ufosLasVegas['mes'] == '03', :]

# Paso duration a un tipo numerico.
ufosLasVegas['duration'] = ufosLasVegas['duration (seconds)'].apply(pd.to_numeric, errors = 'coerse')

# Elimino la columnas que ya no me sirven.
del ufosLasVegas['mes']
del ufosLasVegas['anio']
del ufosLasVegas['datetime']
del ufosLasVegas['duration (seconds)']

# Ordeno por  duration.
ufosLasVegas = ufosLasVegas.sort_values(by = 'duration', ascending = False)

ufosLasVegas.head(10)

# RTA: Verdadero - Para el anio 2004 y mes de Marzo el avistaje con mayor duracion fue en Las Vegas, Nevada.
```

Luego de analizar los datos publicados en: https://www.kaggle.com/uciml/student-alcohol-consumption Indicar cuál de las siguientes afirmaciones es correcta:	 
 	
a) Existen mas alumnos de Portugués en la escuela Mousinho da Silveira que en la Gabriel Pereira.
 	
b) Los alumnos cuyos padres y madres trabajan en igual área (por ejemplo ambos trabajan en educación) tienen mejor promedio de calificaciones que aquellos cuyo padres trabajan en áreas distintas (por ej la madre trabaja en servicios y el padre en salud).
 	
c) Los alumnos de la escuela Gabriel Pereira tienen mejor desempeño que los de la escuela Mousinho da Silveira, tanto para Matemática como para Portugués.
 	
d) Hay mas registros de alumnos de Matemática que de Portugués.

```python
mathematicCourse = pd.read_csv('Data/student-mat.csv')
portugueseCourse = pd.read_csv('Data/student-por.csv')
mathematicCourse
```

```python
# a) Existen mas alumnos de Portugués en la escuela Mousinho da Silveira que en la Gabriel Pereira.
school = portugueseCourse[['school']]
school['count'] = 1
school = school.groupby(['school'], as_index = False).count()
school

# RTA: Falsa - Existen 423 alumnos en la escuela Gabriel Pereira y solo 226 en la escuela Mousinho da Silveira.
```

```python
# b) Los alumnos cuyos padres y madres trabajan en igual área (por ejemplo ambos trabajan en educación) 
#    tienen mejor promedio de calificaciones que aquellos cuyo padres trabajan en áreas distintas 
#    (por ej la madre trabaja en servicios y el padre en salud).
parentsJobsPor = portugueseCourse[['Mjob', 'Fjob', 'G1', 'G2', 'G3']]
parentsJobsPor['sameJob'] = np.where(parentsJobsPor['Mjob'] == parentsJobsPor['Fjob'], "same", "distinct")

del parentsJobsPor['Mjob']
del parentsJobsPor['Fjob']

parentsJobsPor = parentsJobsPor.groupby(['sameJob'], as_index = False).mean()

parentsJobsPor
```

```python
parentsJobsMat = mathematicCourse[['Mjob', 'Fjob', 'G1', 'G2', 'G3']]
parentsJobsMat['sameJob'] = np.where(parentsJobsMat['Mjob'] == parentsJobsMat['Fjob'], "same", "distinct")

del parentsJobsMat['Mjob']
del parentsJobsMat['Fjob']

parentsJobsMat = parentsJobsMat.groupby(['sameJob'], as_index = False).mean()

parentsJobsMat
```

```python
# RTA: Falso - Se puede ver que en el 'G1' del curso de matematicas los alumnos con padres cuyos empleos son distintos
#              Consiguieron mejor promedio que aquellos cuyos padres tienen igual empleo.
```

```python
# c) Los alumnos de la escuela Gabriel Pereira tienen mejor desempeño que los de la escuela Mousinho da Silveira, 
#    tanto para Matemática como para Portugués.
schoolPor = portugueseCourse[['school', 'G1', 'G2', 'G3']]
schoolPor = schoolPor.groupby(['school'], as_index = False).mean()
schoolPor
```

```python
schoolMat = mathematicCourse[['school', 'G1', 'G2', 'G3']]
schoolMat = schoolMat.groupby(['school'], as_index = False).mean()
schoolMat
```

```python
# RTA: Verdadero - En los 3 periodos analizados, para las 2 materias analizadas, los alumnos de la escuela 
#                  Gabriel Pereira tienen mejores calificaciones que los alumnos de la escuela Silveira.
```

6.  Luego de analizar los datos publicados en: https://www.kaggle.com/NUFORC/ufo-sightings Indicar cuál de las siguientes afirmaciones es correcta:	 

a) Existen mas reportes de ufos en forma de cruz que de diamante.

b) Existen 143 registros de avistajes en Canadá durante el año 2010

c) Entre los avistajes registrados en Washington, DC, las formas mas populares son luz, esfera y bola de fuego.

d) Los avistajes en forma de bola de fuego tienen un tiempo promedio de duración de aproximadamente el doble que el de los de forma de huevo.

```python
# a) Existen mas reportes de ufos en forma de cruz que de diamante.
ufosShape = ufosDf[['shape']]
ufosShape['count'] = 1
ufosShape = ufosShape.groupby(['shape'], as_index = False).count()
ufosShape = ufosShape.sort_values(by = 'count', ascending = False)
ufosShape

# RTA: Falso - Existen 1178 reportes de ovnis con forma de diamantes y 233 con forma de cruz.
```

```python
# b) Existen 143 registros de avistajes en Canadá durante el año 2010.

# Me quedo solo con los avistajes de canada.
ufosCanada = ufosDf[['country', 'datetime']]
ufosCanada = ufosCanada.loc[ufosCanada['country'] == 'ca', :]

# Formateo la fecha y tomo el anio.
ufosCanada['datetime'] = pd.to_datetime(ufosCanada['datetime'], format='%m/%d/%Y %H:%M', errors='coerce')
ufosCanada.dropna(subset = ['datetime'], inplace = True)
ufosCanada['anio'] = ufosCanada['datetime'].apply(lambda x: x.strftime('%Y'))

# Me quedo solo con los avistajes de 2010.
ufosCanada = ufosCanada.loc[ufosCanada['anio'] == '2010', :]

len(ufosCanada)

# RTA: Falso - Existen 137 avistajes en Canada en 2010.
```

```python
# c) Entre los avistajes registrados en Washington, DC, las formas mas populares son luz, esfera y bola de fuego.

# Me quedo solo con las columnas que me interesan y filtro por Washington, DC.
ufosWashington = ufosDf[['city', 'state', 'shape']]
ufosWashington = ufosWashington.loc[ufosWashington['state'] == 'dc', :]
ufosWashington = ufosWashington.loc[ufosWashington['city'].str.contains('washington'), :]

# Elimino las columnas innecesarias y agrego un contador.
del ufosWashington['city']
del ufosWashington['state']
ufosWashington['count'] = 1

# Agrupo y ordeno por forma de avistaje.
ufosWashington = ufosWashington.groupby(['shape'], as_index = False).count()
ufosWashington = ufosWashington.sort_values(by = 'count', ascending = False)

ufosWashington

# RTA: Verdadero.
```
